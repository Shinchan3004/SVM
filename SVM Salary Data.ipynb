```python
string_columns = ['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary']
```

```python
label_encoder = preprocessing.LabelEncoder()
for i in string_columns:
    train[i]=label_encoder.fit_transform(train[i])
    test[i]=label_encoder.fit_transform(test[i])
```

```python
train
```

```python
x_train = train.iloc[:,0:13]
y_train = train.iloc[:,-1]
x_test = test.iloc[:,0:13]
y_test = test.iloc[:,-1]
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = norm_func(x_train)
x_test = norm_func(x_test)
```

## Grid Search CV

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[0.1,0.2,0.3,0.5],'C':[15,10,1,0.1,0.001],'random_state':[40] }]
```

```python
gsv = RandomizedSearchCV(clf,param_grid,cv=5)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 15, gamma = 0.2, kernel = 'rbf')
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
